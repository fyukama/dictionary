## Linear Algebra

Vectors and matrices are the language of neural networks.

```
v1=np.array([2,1]); v2=np.array([1,2])
plt.figure(figsize=(5,5))
plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,label='v1')
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,label='v2')
plt.xlim(0,3); plt.ylim(0,3); plt.grid(); plt.title("Vectors"); plt.show()
```

## Calculus

Gradients tell us which direction reduces error the fastest.

```
x=np.linspace(-3,3,300); y=x**2
plt.plot(x,y)
plt.scatter([1],[1])
plt.annotate("Gradient points uphill",xy=(1,1),xytext=(1.5,4),arrowprops=dict(arrowstyle='->'))
plt.grid(); plt.show()
```

## Neuron

## Activation Functions

Nonlinearities allow networks to learn complex patterns.

```
x=np.linspace(-5,5,400)
plt.plot(x,1/(1+np.exp(-x)),label='Sigmoid')
plt.plot(x,np.maximum(0,x),label='ReLU')
plt.plot(x,np.tanh(x),label='Tanh')
plt.legend(); plt.grid(); plt.show()
```

## Forward Propagation

Information flows from inputs to outputs.

## Loss Functions

Loss measures prediction error.

```
actual=np.array([1,2,3,4]); pred=np.array([1.1,1.8,2.9,4.2])
print("MSE:",np.mean((actual-pred)**2))
```

## Gradient Descent

Parameters are updated to reduce loss.

```
x=np.linspace(-4,4,300); y=x**2
plt.plot(x,y)
pts=[2,1,0.5,0.2,0]
plt.scatter(pts,[p**2 for p in pts])
plt.grid(); plt.show()
```

## Backpropagation

Gradients are propagated backward using the chain rule.

## Training Loop

Forward → Loss → Backprop → Update.

## Overfitting

Compare training and validation performance.

```
epochs=np.arange(1,11)
train=np.exp(-epochs/4)
val=np.array([0.9,0.7,0.6,0.55,0.56,0.58,0.62,0.68,0.75,0.84])
plt.plot(epochs,train,label="Training")
plt.plot(epochs,val,label="Validation")
plt.legend(); plt.grid(); plt.show()
```

## Attention

Different input tokens contribute unequally.

```
tokens=["The","cat","sat","mat"]
weights=[0.1,0.45,0.15,0.3]
plt.bar(tokens,weights)
plt.ylabel("Attention Weight")
plt.show()
```

## Embeddings

Objects are represented as dense vectors.

```
points=np.array([[1,2],[2,2.5],[4,4],[4.5,4.2]])
labels=["cat","dog","car","truck"]
plt.scatter(points[:,0],points[:,1])
for p,l in zip(points,labels):
    plt.text(p[0]+0.05,p[1]+0.05,l)
plt.title("Toy Embedding Space")
plt.grid(); plt.show()
```
